````python
import pandas as pd
import pyarrow.parquet as pq
import os
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from pandas.tseries.holiday import USFederalHolidayCalendar as calander
import warnings

warnings.filterwarnings("ignore")

pd.set_option("display.max_columns", 30)
color = sns.color_palette("tab10")
````

````python
trips = pq.read_table('/content/drive/MyDrive/Bigdata/july/yellow_tripdata_2023-07.parquet')
trips = trips.to_pandas()
data = trips
````

````python
data
````

````python
# ดึงข้อมูลวันและเวลา จากนั้นคำนวณระยะเวลาของการเดินทาง
# แปลงคอลัมน์วันที่และเวลาให้เป็นชนิด datetime
data['tpep_pickup_datetime'] = pd.to_datetime(data['tpep_pickup_datetime'])
data['tpep_dropoff_datetime'] = pd.to_datetime(data['tpep_dropoff_datetime'])

# สร้างคอลัมน์วันที่ (date) โดยลบส่วนเวลาจาก datetime
data['date'] = data['tpep_pickup_datetime'].dt.normalize()

# สร้างคอลัมน์เวลาที่แยกเฉพาะชั่วโมงจาก datetime
data['time'] = data['tpep_pickup_datetime'].dt.hour

# สร้างคอลัมน์ชื่อวันในสัปดาห์จากวันที่
data['weekday'] = data['date'].dt.day_name()

# คำนวณระยะเวลาการเดินทางโดยลบเวลาเริ่มต้นจากเวลาสิ้นสุด
data['duration'] = data['tpep_dropoff_datetime'] - data['tpep_pickup_datetime']

# แปลงระยะเวลาเป็นหน่วยนาที
data['duration'] = data['duration'] / np.timedelta64(1, 'm')
````

````python
# การระบุวันหยุด
cal = calander()
# ดึงวันหยุดในช่วงเวลาที่กำหนด
holidays = cal.holidays(start='2023-01-01', end='2023-12-31')
# สร้างคอลัมน์ใหม่ holiday เพื่อระบุว่าแต่ละวันที่ในข้อมูลเป็นวันหยุดหรือไม่
data['holiday'] = data['date'].isin(holidays)
# สร้างคอลัมน์ใหม่ week เพื่อระบุวันในสัปดาห์ (0=Monday, 6=Sunday)
data['week'] = data['date'].dt.dayofweek
# กำหนดประเภทวัน (day_type) เป็น weekend
data.loc[data['week'] >= 5, 'day_type'] = "weekend"
# กำหนดประเภทวัน (day_type) เป็น workday
data.loc[data['week'] < 5, 'day_type'] = "workday"
data.loc[data['holiday'] == True, 'day_type'] = "holiday"
# ลบคอลัมน์ที่ไม่จำเป็น (holiday และ week)
data = data.drop(['holiday', 'week'], axis=1)
````

````python
# คำนวณเปอร์เซ็นต์ทิปโดยใช้จำนวนทิปและจำนวนทั้งหมด
data['tip_percent'] = (data.tip_amount / data.total_amount) * 100
````

````python
# จัดประเภทช่วงเวลาของการเดินทาง

def time_slots(x):
    if x in range(6,12):
        return 'Morning'
    elif x in range(12,17):
        return 'Afternoon'
    elif x in range(17,22):
        return 'Evening'
    else:
        return 'Late Night'

data['time_desc'] = data['time'].apply(time_slots)
````

````python
# ลบฟีเจอร์ที่เราไม่สนใจออก
data = data.drop(['tpep_pickup_datetime', 'tpep_dropoff_datetime', 'store_and_fwd_flag', 'extra', 'mta_tax', 'improvement_surcharge'], axis=1)
````

````python
# เพื่อสร้างกราฟแสดงการกระจายของระยะทางการเดินทาง
sns.distplot(data[(data['trip_distance'] > 0) & (data['trip_distance'] <= 50)]['trip_distance'])
````

````python
# สร้างกราฟแสดงการกระจายของระยะเวลา
sns.distplot(data[(data['duration'] > 0) & (data['duration'] < 120)]['duration'])
````

````python
# กราฟแสดงการกระจายของจำนวนค่าโดยสาร
sns.distplot(data[(data['fare_amount'] >= 2.5)&(data['fare_amount'] <= 150)]['fare_amount'])
````

````python
# กราฟแสดงการกระจายของจำนวนทิป
sns.distplot(data.tip_amount)
````

````python
data.info()
````

````python
# ลบข้อมูลที่ไม่ถูกต้อง
# จำนวนผู้โดยสาร == 0
data = data[(data['passenger_count'] > 0) & (data['passenger_count'] < 7)]

# ระยะทางการเดินทาง <= 0
data = data[(data['trip_distance'] > 0) & (data['trip_distance'] <= 100)]

# ยกเว้นระยะเวลาที่นานกว่า 180 นาที
data = data[(data['duration'] > 0) & (data['duration'] <= 180)]

# วิธีการชำระเงินที่ไม่ใช่เงินสดและบัตรเครดิต
data = data[(data['payment_type'] != 3) & (data['payment_type'] != 4) & (data['payment_type'] != 5) & (data['payment_type'] != 6)]

# ยกเว้นกรณีที่จำนวนค่าโดยสารน้อยกว่า 2.5
data = data[(data['fare_amount'] >= 2.5) & (data['fare_amount'] <= 250)]

# ลบการเดินทางที่เปอร์เซ็นต์ทิปมากกว่า 50%
data = data[(data['tip_percent'] >= 0) & (data['tip_percent'] <= 50)]

# รหัสสถานที่รับและส่งไม่อยู่ในโซนแท็กซี่ NYC
data = data[(data['PULocationID'] >= 1) & (data['PULocationID'] <= 263) & (data['DOLocationID'] >= 1) & (data['DOLocationID'] <= 263)]
````

````python
# # แปลงจำนวนผู้โดยสารเป็นสามระดับ
def passenger_convert(row):
    if (row >= 1) & (row <= 2):
        val = "small"
    elif (row > 2) & (row < 5):
        val = "medium"
    else:
        val = "high"
    return val

data['passenger_count'] = data['passenger_count'].apply(passenger_convert)
````

````python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import contextily as ctx
import numpy as np

# คำนวณจำนวนการรับ (pickup) และส่ง (dropoff)
def create_spatial_visualization(all_data, taxi_zones_gdf):
    """
    สร้างการแสดงผลเชิงพื้นที่ของรูปแบบการรับและส่ง

    พารามิเตอร์:
    -----------
    all_data : pandas DataFrame
        ข้อมูลการเดินทางโดยแท็กซี่ที่รวมกัน
    taxi_zones_gdf : GeoDataFrame
        โซนแท็กซี่พร้อมข้อมูลทางเรขาคณิต
    """
    # คำนวณจำนวนสำหรับแต่ละโซน
    pu_counts = all_data['PULocationID'].value_counts().reset_index()
    pu_counts.columns = ['LocationID', 'pickup_count']

    do_counts = all_data['DOLocationID'].value_counts().reset_index()
    do_counts.columns = ['LocationID', 'dropoff_count']

    #  สร้าง GeoDataFrames ที่มีจำนวน
    pu_zone_gdf = taxi_zones_gdf.merge(pu_counts, on='LocationID', how='left')
    do_zone_gdf = taxi_zones_gdf.merge(do_counts, on='LocationID', how='left')

    # แทนที่ค่า NaN ด้วย 0
    pu_zone_gdf['pickup_count'] = pu_zone_gdf['pickup_count'].fillna(0)
    do_zone_gdf['dropoff_count'] = do_zone_gdf['dropoff_count'].fillna(0)

     # แปลงเป็น EPSG:3857 สำหรับการฉายแผนที่เว็บ Mercator
    pu_zone_gdf = pu_zone_gdf.to_crs(epsg=3857)
    do_zone_gdf = do_zone_gdf.to_crs(epsg=3857)

     # สร้างการแสดงผล
    fig, axs = plt.subplots(ncols=2, figsize=(20, 10))
    fig.subplots_adjust(bottom=0.1, top=0.9, left=0.09, right=0.8, wspace=0.02, hspace=0.02)


    # แสดงรูปแบบการรับ
    pu_plot = pu_zone_gdf.plot(column='pickup_count',
                              ax=axs[0],
                              cmap='Blues',
                              legend=False,
                              linewidth=0.09)

      # แสดงรูปแบบการส่ง
    do_plot = do_zone_gdf.plot(column='dropoff_count',
                              ax=axs[1],
                              cmap='Blues',
                              legend=False,
                              linewidth=0.09)

   # เพิ่มแผนที่ฐาน
    ctx.add_basemap(axs[0], crs=pu_zone_gdf.crs.to_string())
    ctx.add_basemap(axs[1], crs=do_zone_gdf.crs.to_string())

    # ตั้งชื่อกราฟ
    axs[0].set_title('Pickup Patterns')
    axs[1].set_title('Dropoff Patterns')

    # ลบแกน
    axs[0].set_axis_off()
    axs[1].set_axis_off()

    # เพิ่มแถบสี
    norm = plt.Normalize(vmin=min(pu_zone_gdf['pickup_count'].min(),
                                do_zone_gdf['dropoff_count'].min()),
                        vmax=max(pu_zone_gdf['pickup_count'].max(),
                                do_zone_gdf['dropoff_count'].max()))

    sm = plt.cm.ScalarMappable(cmap='Blues', norm=norm)
    cb = fig.colorbar(sm, ax=axs, shrink=0.72, orientation="vertical", pad=0.005)
    cb.ax.set_ylabel('Trip Count')

    plt.show()

    return pu_zone_gdf, do_zone_gdf

# ตัวอย่างการใช้งาน
if __name__ == "__main__":
 # อ่านไฟล์ shapefile
    taxi_zones = gpd.read_file('/content/shapefile_extracted/taxi_zones.shp')
    taxi_zones = taxi_zones.to_crs(epsg=4326)  # Ensure it's in WGS84

    # สร้างการแสดงผล
    pu_zone_gdf, do_zone_gdf = create_spatial_visualization(summer_data, taxi_zones)

     # แสดงโซนการรับและส่ง 5 อันดับแรก
    print("\nTop 5 Pickup Zones:")
    print(pu_zone_gdf.nlargest(5, 'pickup_count')[['zone', 'pickup_count']])

    print("\nTop 5 Dropoff Zones:")
    print(do_zone_gdf.nlargest(5, 'dropoff_count')[['zone', 'dropoff_count']])
````

````python
# จัดกลุ่มและนับจำนวนการเดินทางตามวันที่และสภาพอากาศ

winter_prcp = pd.DataFrame(winter_data.groupby(['date', 'precip']).size()).reset_index()
summer_prcp = pd.DataFrame(summer_data.groupby(['date', 'precip']).size()).reset_index()
winter_prcp.columns = ['date', 'precip', 'pickup_freq']
summer_prcp.columns = ['date', 'precip', 'pickup_freq']

winter_awnd = pd.DataFrame(winter_data.groupby(['date', 'windspeed']).size()).reset_index()
summer_awnd = pd.DataFrame(summer_data.groupby(['date', 'windspeed']).size()).reset_index()
winter_awnd.columns = ['date', 'windspeed', 'pickup_freq']
summer_awnd.columns = ['date', 'windspeed', 'pickup_freq']

winter_snow = pd.DataFrame(winter_data.groupby(['date', 'snow']).size()).reset_index()
summer_snow = pd.DataFrame(summer_data.groupby(['date', 'snow']).size()).reset_index()
winter_snow.columns = ['date', 'snow', 'pickup_freq']
summer_snow.columns = ['date', 'snow', 'pickup_freq']

winter_snwd = pd.DataFrame(winter_data.groupby(['date', 'snowdepth']).size()).reset_index()
summer_snwd = pd.DataFrame(summer_data.groupby(['date', 'snowdepth']).size()).reset_index()
winter_snwd.columns = ['date', 'snowdepth', 'pickup_freq']
summer_snwd.columns = ['date', 'snowdepth', 'pickup_freq']
````

````python
# # กำหนดฟังก์ชันช่วยเหลือสำหรับการฟิตพหุนามอย่างปลอดภัย
def safe_polyfit(x, y):
     """ทำการฟิตพหุนามอย่างปลอดภัยโดยมีการจัดการข้อผิดพลาด"""
    try:
         # ลบค่าที่ไม่เป็นตัวเลข (NaN) ออก
        mask = ~(np.isnan(x) | np.isnan(y))
        x_clean = x[mask]
        y_clean = y[mask]

        # ตรวจสอบว่ามีจุดข้อมูลเพียงพอหรือไม่
        if len(x_clean) < 2:
            return None, None

        # ทำการฟิตพหุนาม
        z = np.polyfit(x_clean, y_clean, 1)
        p = np.poly1d(z)
        return z, p
    except Exception as e:
        print(f"Error in polynomial fitting: {e}")
        return None, None

# Create the plot
fig, axs = plt.subplots(2, 2, figsize=(12, 10), sharey=True)

# กราฟผลกระทบของปริมาณน้ำฝน (top left)
for data, season, color in [(winter_prcp, 'Winter', 'b'), (summer_prcp, 'Summer', 'r')]:
    z, p = safe_polyfit(data['precip'], data['pickup_freq'])
    if p is not None:
        axs[0, 0].plot(data['precip'], p(data['precip']), c=color)
    axs[0, 0].scatter(data['precip'], data['pickup_freq'], c=color, label=season, alpha=0.5)
axs[0, 0].set_xlabel('Precipitation (inches)')
axs[0, 0].set_ylabel('Number of pickups')
axs[0, 0].set_title('Precipitation Effect')
axs[0, 0].grid(True, alpha=0.3)


# กราฟผลกระทบของความเร็วลม (top right)
for data, season, color in [(winter_awnd, 'Winter', 'b'), (summer_awnd, 'Summer', 'r')]:
    z, p = safe_polyfit(data['windspeed'], data['pickup_freq'])
    if p is not None:
        axs[0, 1].plot(data['windspeed'], p(data['windspeed']), c=color)
    axs[0, 1].scatter(data['windspeed'], data['pickup_freq'], c=color, label=season, alpha=0.5)
axs[0, 1].set_xlabel('Avg. Daily Wind Speed (mph)')
axs[0, 1].set_title('Wind Speed Effect')
axs[0, 1].grid(True, alpha=0.3)

# กราฟผลกระทบของหิมะ (bottom left)
z, p = safe_polyfit(winter_snow['snow'], winter_snow['pickup_freq'])
if p is not None:
    axs[1, 0].plot(winter_snow['snow'], p(winter_snow['snow']), c='b')
axs[1, 0].scatter(winter_snow['snow'], winter_snow['pickup_freq'], c='b', label='Winter', alpha=0.5)
axs[1, 0].set_xlabel('Snow (inches)')
axs[1, 0].set_ylabel('Number of pickups')
axs[1, 0].set_title('Snow Effect')
axs[1, 0].grid(True, alpha=0.3)

# กราฟผลกระทบของความลึกของหิมะ (bottom right)
z, p = safe_polyfit(winter_snwd['snowdepth'], winter_snwd['pickup_freq'])
if p is not None:
    axs[1, 1].plot(winter_snwd['snowdepth'], p(winter_snwd['snowdepth']), c='b')
axs[1, 1].scatter(winter_snwd['snowdepth'], winter_snwd['pickup_freq'], c='b', label='Winter', alpha=0.5)
axs[1, 1].set_xlabel('Snow Depth (inches)')
axs[1, 1].set_title('Snow Depth Effect')
axs[1, 1].grid(True, alpha=0.3)

# Add legend to the first subplot
axs[0, 0].legend(loc='lower right')

# Add main title
plt.suptitle('Weather Conditions vs Pickup Frequency by Day', fontsize=14, y=1.02)

# Adjust layout
plt.tight_layout()

# Display the plot
plt.show()
````

````python
# Read the CSV files
winter_data = pd.read_csv('/content/drive/MyDrive/Bigdata/taxi_data_processed1.csv')
summer_data = pd.read_csv('/content/drive/MyDrive/Bigdata/taxi_data_processed7.csv')

# สามารถเพิ่มคอลัมน์เพื่อแยกความแตกต่างระหว่างข้อมูลฤดูหนาวและฤดูร้อนได้
winter_data['season'] = 'winter'
summer_data['season'] = 'summer'

# รวมข้อมูลฤดูหนาวและฤดูร้อนหากจำเป็น
all_data = pd.concat([winter_data, summer_data], ignore_index=True)
````

````python
# กำหนดชื่อคอลัมน์ที่ต้องการวิเคราะห์
col = ['trip_distance', 'fare_amount', 'tip_amount', 'total_amount', 'duration', 'tolls_amount', 'precip', 'snow', 'windspeed', 'snowdepth', 'avg_temp']
# กำหนดชื่อป้ายสำหรับการแสดงผล
label = ['Distance', 'Fare', 'Tip', 'Total$', 'Duration', 'Tolls', 'PRCP', 'SNOW', 'AWND', 'SNWD', 'TAVG']

# วาดกราฟความสัมพันธ์ (correlation plot)
fig, ax = plt.subplots(figsize=(16, 5))  # กำหนดขนาดของแผนภูมิ
corr_matrix = all_data[col].corr()  # คำนวณค่าความสัมพันธ์
sns.heatmap(corr_matrix, annot=True, ax=ax, cbar_kws={'label': 'Correlation Coefficient'}) # สร้าง heatmap
ax.set_xticklabels(label)
ax.set_yticklabels(label)
ax.set_title("Correlation Between Continuous Attributes")
plt.show()
````

````python
# เก็บเฉพาะการชำระเงินด้วยบัตร เนื่องจากทิปจากการชำระเงินด้วยเงินสดจะไม่ถูกบันทึก

sample = all_data[all_data['payment_type'] == 1].sample(frac=0.05, replace=True, random_state=30034).reset_index(drop=True)
````

````python
# กำหนดชื่อของคอลัมน์ที่จะเลือกในตัวแปร
COL = ['passenger_count', 'fare_amount', 'tip_amount', 'tolls_amount',
       'season', 'day_type', 'time_desc', 'PUborough', 'DOborough']
sample_filtered = sample.loc[:, COL].reset_index(drop=True)
````

````python
# Split data to train and test
y = sample_filtered['tip_amount']
X_train, X_test, y_train, y_test = train_test_split(sample_filtered, y, test_size=0.3, random_state=0)
````

````python
# แบ่งข้อมูลเป็นชุดฝึก (train) และชุดทดสอบ (test)
y = sample_filtered['tip_amount']
X_train, X_test, y_train, y_test = train_test_split(sample_filtered, y, test_size=0.3, random_state=0)
````

````python
# โมเดลฐานซึ่งประกอบด้วยแค่ค่าตัดที่ (intercept) เท่านั้น

baseline = ols(formula='tip_amount ~ 1', data=X_train).fit()
print(baseline.summary())
````

````python

base_train_pred = baseline.predict(X_train) #  ใช้โมเดลฐาน (baseline) เพื่อทำนายค่าทิปจากข้อมูลการฝึก (X_train) และเก็บผลลัพธ์การทำนายไว้ในตัวแปร base_train_pred
base_test_pred = baseline.predict(X_test) # base_test_pred = baseline.predict(X_test): ใช้โมเดลฐานเพื่อทำนายค่าทิปจากข้อมูลการทดสอบ (X_test) และเก็บผลลัพธ์ไว้ในตัวแปร base_test_pred

# คำนวณค่า RMSE (Root Mean Squared Error) สำหรับข้อมูลการฝึก โดยใช้ฟังก์ชัน mean_squared_error ซึ่งจะเปรียบเทียบค่าทิปจริง (y_train)
# กับค่าทิปที่ทำนายได้ (base_train_pred) และตั้งค่า squared=False เพื่อให้ได้ค่า RMSE แทนที่จะเป็น MSE (Mean Squared Error)
train_rmse = mean_squared_error(y_train, base_train_pred, squared=False)

# คำนวณค่า RMSE สำหรับข้อมูลการทดสอบในลักษณะเดียวกับที่ใช้กับข้อมูลการฝึก โดยเปรียบเทียบค่าทิปจริง (y_test) กับค่าทิปที่ทำนายได้ (base_test_pred)
test_rmse = mean_squared_error(y_test, base_test_pred, squared=False)

print("Train RMSE:", train_rmse)
print("Test RMSE:", test_rmse)
````

````python
# ปรับโมเดล
# ใช้ฟังก์ชันจากไลบรารี Statsmodels เพื่อสร้างโมเดลการถดถอยเชิงเส้น (Ordinary Least Squares Regression)

full_model = ols(formula='tip_amount ~ C(passenger_count) + fare_amount + tolls_amount + season + time_desc + day_type + PUborough + DOborough', data=X_train).fit()
print(full_model.summary())
````

````python
# ทำนายและประเมินผล
full_train_pred = full_model.predict(X_train)
full_test_pred = full_model.predict(X_test)
train_rmse = mean_squared_error(y_train, full_train_pred, squared=False)
test_rmse = mean_squared_error(y_test, full_test_pred, squared=False)
train_r2 = r2_score(y_train, full_train_pred)
test_r2 = r2_score(y_test, full_test_pred)

print("Full model")
print("Train RMSE:", train_rmse)
print("Test RMSE:", test_rmse)
print("Train R2", train_r2)
print("Test R2", test_r2)
````
